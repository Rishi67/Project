```julia
using JLD2, FileIO, Knet

loadfile = "/Users/rishi/Project/GAN/invert/models/0040.jld2"

function convert_weights(w, atype=Array{Float32})
    w0 = map(wi->convert(atype, wi), w)
    w1 = convert(Array{Any}, w0)
end


function convert_moments(moments,atype=Array{Float32})
    clone = map(mi->bnmoments(), moments)
    for k = 1:length(clone)
        if moments[k].mean != nothing
            clone[k].mean = convert(atype, moments[k].mean)
        end

        if moments[k].var != nothing
            clone[k].var = convert(atype, moments[k].var)
        end
    end
    return convert(Array{Any,1}, clone)
end
```

```julia
function leaky_relu_inverse(x,alpha=0.02)
    pos = max(0,x)
    neg = min(0,x)/alpha
    return pos + neg
end

function inverse_batchnorm(x,moments,params)
    (N,m) = size(x)
    
    scale = params[1:N]
    bias = params[N+1:end]
    
    m = moments.mean
    v = moments.var
    
    for i = 1:N
        x[i,:] = (((x[i,:] .- bias[i]) .* (1/scale[i])).*sqrt(v[i]+1e-5)).+m[i]
    end
    
    return x
end
        
function inverse_glayer(w,x,m)
    x = leaky_relu_inverse.(x)
    x = inverse_batchnorm(x,m,w[2])
    x = pinv(w[1])*x
    return x
end
```

```julia
function gnet_inverse2(wg,mg,y)
    y = reshape(y, 784,:)
    println(size(y))
    y1 = atanh.(y)
    println(size(y1))
    println(size(wg[end]))
    y2 = y1 .- wg[end]
    y3 = pinv(wg[9])*y2
    y4 = inverse_glayer(wg[7:8],y3,mg[4])
    y5 = inverse_glayer(wg[5:6],y4,mg[3])
    y6 = inverse_glayer(wg[3:4],y5,mg[2])
    y7 = inverse_glayer(wg[1:2],y6,mg[1])
    
    return y7
end
```

```julia
function load_digit_data(digit,nx=28,ny=28,nrep=1000)
    file = "/Users/rishi/Project/GAN/data/data"*string(digit)*".dms"
    fp = open(file, "r")
    x = read(fp, UInt8, (nx,ny,1,nrep)) # what is the type of x0?
    close(fp)
    
   return x
end

function load_class_data(num_digits)
    digits = collect(0:num_digits)
    num_digits = length(digits)
    x = load_digit_data.(digits)
    
    xtrain = x[1]
    for i=2:num_digits
        xtrain = cat(4,xtrain, x[i])
    end
    
    xtst = xtrain[:,:,:,901:1000]
    for i=1:num_digits-1
        xtst = cat(4, xtst, xtrain[:,:,:,(i*1000)+901:(i+1)*1000])
    end
        
    ytst = ones(UInt8,100*num_digits)
    ytrn = ones(UInt8,1000*num_digits)
    
    return xtrain,ytrn,xtst,ytst
end
```

```julia
A,b,c,d = load_class_data(9)
A = reshape(A, 784,:)
A = convert(Array{Float64}, A)
```

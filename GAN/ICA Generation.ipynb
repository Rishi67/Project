```julia
using Plots, Interact, JLD, Colors
```

```julia
D = load("icaDigits.jld");
X = D["X"];
Q = D["Q"];
```

```julia
hist = zeros(784,1000,10)
for i = 1:10
    s = ((i-1)*1000)+1
    e = s+999
    hist[:,:,i] = X[:,s:e]
end
```

```julia
for i = 1:10
    for j = 1:784
        hist[j,:,i] = sort(hist[j,:,i])
    end
end
```

```julia
x = ceil.(1000*rand(784,1))
for i = 1:784
    x[i] = hist[i,Int(x[i]),3]
end
```

```julia
im = reshape(Q*x,28,28)
heatmap(im)
```
